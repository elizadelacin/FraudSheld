```python
import pandas as pd 
import numpy as np 

transaction = pd.read_csv("../data/raw/transaction.csv")
identity = pd.read_csv("../data/raw/identity.csv")

print("Transaction shape:", transaction.shape)
print("Identity shape:", identity.shape)
```

```python
df = pd.merge(transaction, identity, on='TransactionID', how='left')
print(df.shape)
print(df['isFraud'].value_counts())
```

```python
nan_percent = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
print(nan_percent[nan_percent > 0])
```

```python
cols_to_drop = nan_percent[nan_percent > 80].index
df = df.drop(columns=cols_to_drop)
print(df.shape)
print(len(cols_to_drop))
```

```python
numerical_cols = df.select_dtypes(include=["float64", "int64"]).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

categorical_cols = df.select_dtypes(include="object").columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")
```

```python
print(df.isna().sum().sum())
```

```python
df.to_parquet('../data/processed/clean_data.parquet', index=False)

```
